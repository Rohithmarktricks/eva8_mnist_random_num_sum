### Load the required libraries

```python
# import the essential libraries
import numpy as np
import torch
from torchvision import datasets
import torchvision.transforms as transforms
from torchvision.utils import make_grid
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Check if the "GPU/CUDA" is available, If Yes, make the default device as GPU (for faster calculations)
use_cuda = torch.cuda.is_available()
device = torch.device("cuda" if use_cuda else "cpu")
print(device)
```

### Data Loading

```python
torch.manual_seed(1) # To make sure that the same splitting/shuffling is done everytime.
batch_size = 128 #batch size of 128 (images) to be loaded at once.
kwargs = {'num_workers': 1, 'pin_memory': True} if use_cuda else {} # keyword arguments that can be used while loading the data.


# DataLoader API to get the datasets loaded (as a generator/iterable). This way
# train_loader and test_loader can be used to load the data in batches. 
train_loader = torch.utils.data.DataLoader(
    datasets.MNIST('../data', train=True, download=True,
                    transform=transforms.Compose([
                        transforms.ToTensor(),
                        transforms.Normalize((0.1307,), (0.3081,))
                    ])),
    batch_size=batch_size, shuffle=True, **kwargs)

# test_loader to load the data(images, labels) in batches
test_loader = torch.utils.data.DataLoader(
    datasets.MNIST('../data', train=False, transform=transforms.Compose([
                        transforms.ToTensor(),
                        transforms.Normalize((0.1307,), (0.3081,))
                    ])),
    batch_size=batch_size, shuffle=True, **kwargs)
```

### Plotting the MNIST Images and Labels

```python
for im, la in train_loader:
    print('images shape: ', im.shape)
    plt.figure(figsize=(16,8))
    plt.axis('off')
    plt.imshow(make_grid(im, nrow=10).permute(1,2,0))
    break
```

### Convolutional Neural Network

```python
# Convolutional Neural Network

class Net2(nn.Module):
    '''
    This Net2 architecture is used to solve the assignment.
    Contains the following layers:
        7 convolutional layers
        2 MaxPooling layers
        3 Linear/Fully connected layers
    Inputs:
        Image : 1x28x28 (MNIST Image)
        RandomNumber: 0-9
    
    Outputs:
        label: Label of the MNIST Image
        sum_output: Sum of the predicted label of the MNIST Image and the random number.
    
    '''
    def __init__(self, num_classes):
        super(Net2, self).__init__()
        self.conv1 = nn.Conv2d(1, 32, 3, padding=1) #input -1 Output 32; kernel_size = 3, padding=1; Image_size=28x28
        self.conv2 = nn.Conv2d(32, 64, 3, padding=1)# input- 32 channels, output=64 channels, kernel_size = 3, padding=1; image_size = 28x28
        self.pool1 = nn.MaxPool2d(2, 2) #Max Pooling of 2x2 ; image_size = 14x14
        self.conv3 = nn.Conv2d(64, 128, 3, padding=1) #input = 64, output=128, kernel_size=3, padding=1; image_size = 14x14
        self.conv4 = nn.Conv2d(128, 256, 3, padding=1) #input = 128, output=256, kernel_size=3, padding=1; image_size = 14x14
        self.pool2 = nn.MaxPool2d(2, 2) #Max Pooling of 2x2; image_size = 7x7
        self.conv5 = nn.Conv2d(256, 512, 3, padding=1) #input = 256, output=512, kernel_size=3, padding=1; image_size = 7x7
        self.conv6 = nn.Conv2d(512, 1024, 3, padding=1) #input = 512, output=1024, kernel_size = 3; padding=1; image_size = 7x7
        self.conv7 = nn.Conv2d(1024, 64, 3, padding=1) #input = 1024, output=64, kernel_size = 3; padding=1; image_size = 7x7
        self.fc1 = nn.Linear(7*7*64+10, 512) #input is the combination of flattened out image (7x7x64) and +10 (one hot encoded) random number
        self.fc2 = nn.Linear(512, num_classes) #output layer for MNIST Image classification
        self.fc3 = nn.Linear(512, 19) # Output layer for summation of MNIST image predicted label and random number

    def forward(self, x, r):
        x = self.pool1(F.relu(self.conv2(F.relu(self.conv1(x)))))
        x = self.pool2(F.relu(self.conv4(F.relu(self.conv3(x)))))
        x = F.relu(self.conv6(F.relu(self.conv5(x))))
        x = F.relu(self.conv7(x))
        x = x.view(-1, 7*7*64) #Flatten the feature maps to a 1-D tensor.
        x = torch.cat((x, r), dim=1) # concatenate the feature map(1D) and the random number (one hot encoded) into a single tensor
        x = F.relu(self.fc1(x))
        label = self.fc2(x) # Label of the MNIST dataset
        sum_output = self.fc3(x) # sum of the MNIST label and the random number.
        return label, sum_output
```

### Loss Functions and the Optimizer

```python
# Define the number of classes
num_classes = 10

# Initialize the model and move it to the GPU.
model = Net2(num_classes).to(device)

# Define the loss function and optimizer
criterion = nn.CrossEntropyLoss() #Loss function of the MNIST image classification.

criterion2 = nn.CrossEntropyLoss() # Loss function of the summation part. This should ideally be a MSE (Mean square Error) 
# as the summation could be thought of a "regression" problem, however, in the problem/assignment statement, we know that the
# random number input is always bounded between 0-9. And hence the output is also bounded, and it going to be an Integer, 
# as it's allowed to use one-hot encoded representation, the entire summation part could be thought of the classification problem.
# Hence, CrossEntropyLoss has been used here.


# Adam Optimizer with the default hyper-parameters has been used.
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# Total number of epoch requried for the training purpose
num_epochs = 10
```

### Training Loop

```python
loss_epochs = []
# Train the model for 10 epochs
for epoch in range(num_epochs):
    # Loop over the data iterator and sample a batch of images and labels
    for images, labels in train_loader:
        # Convert images and labels to tensors
        images = images.reshape(-1, 1, 28, 28).to(device)
        
        # Unsqueeze the labels, to match the dimensions.
        labels_for_loss = labels.unsqueeze(dim=1)
        # Conver the labels to one-hot encoded representation.
        labels_one_hot = F.one_hot(labels)
        # Conver the one-hot encoding to "float" type and move GPU.
        labels_one_hot = labels_one_hot.type(dtype=torch.float64).to(device)
        
        
        # Generate a random number and convert it to a tensor
        random_int = torch.randint(0, 10, (images.size(0), 1))
        # Convert the random integer to one-hot encoded representation.
        random_one_hot = F.one_hot(random_int).squeeze()
        # Move the one-hot encoded random integer to CUDA/GPU device.
        random_one_hot = random_one_hot.to(device)
        
        # We need to get the true label for the summation part. So, we add the true label of the MNIST image the generated
        # random number (0-9)
        sum_for_loss = random_int + labels_for_loss
        # And convert them to One-hot encoded values, and would be used to calculate the Cross entropy Loss.
        sum_for_loss = F.one_hot(sum_for_loss, 19).squeeze().type(dtype=torch.float64).to(device)
        
        # Forward pass
        label_output, sum_output = model(images, random_one_hot)
        # calculate the loss for the MNIST Image Classification
        loss_mnist = criterion(label_output, labels_one_hot)
        # calculate the loss for the MNIST Image label + random integer summation part.
        loss_sum = criterion2(sum_output, sum_for_loss)
        # Add the corresponding loss values.
        loss = loss_mnist + loss_sum
        
        # Backward pass and optimize
        optimizer.zero_grad()
        
        loss.backward()
        optimizer.step()
    loss_epochs.append(loss.item())
    print(f"Loss at Epoch {epoch}: {loss}")
```

### Loss Plot (Training)

```python
plt.plot(list(range(num_epochs)), loss_epochs)
plt.xlabel("Epochs")
plt.ylabel("Loss Value")
plt.title("Loss Plot")
```

### Testing the model

```python
def test_model():
    
    model.eval() # force the model to the evaluation mode
    
    # variables to compute the final Loss and Accuracy
    test_loss = 0
    correct = 0
    correct_sum = 0
    test_sum_loss =0
    
    # Use no_grad() context manager, as we don't want to track for backpropagation!!
    with torch.no_grad():
        # Loop through the test data loader.
        for images, labels in test_loader:
            
            #conver the images to the required shape (batch_size, channels, H, W) and move to the GPU.
            images = images.reshape(-1, 1, 28, 28).to(device)
            # Move the labels to the GPU.
            labels_2 = labels.to(device)
            # Get the one-hot encoded representation of the labels.
            labels2_one_hot = F.one_hot(labels)
            labels2_one_hot= labels2_one_hot.type(dtype=torch.float64).to(device)

            # Generate the random integers between 0-9.
            random2 = torch.randint(0, 10, (images.size(0), 1)).to(device)
            # Get the one-hot encoded representation of the random integers.
            random2_one_hot = F.one_hot(random2).squeeze().to(device)
            
            # Get the MNIST label and the Summation part from the model (as predictions)
            label_output, sum_output = model(images, random2_one_hot)
            # Unsqueeze the labels to add the extra dimension
            labels_to_check = labels_2.unsqueeze(dim=1)
            
            # Get the summation part (MNIST True label + Random Integer value)
            sum_to_check = random2 + labels_to_check
            sum_to_check = sum_to_check.to(device)
            # Convert it to the one-hot encoded representation!
            sum_to_check_one_hot = F.one_hot(sum_to_check, 19).squeeze().type(dtype=torch.float64).to(device)
            
            
            # Calculate the complete loss of the MNIST labels.
            test_loss += criterion(label_output, labels2_one_hot)
            # Calculate the complete loss of the summation part!
            test_sum_loss += criterion(sum_output, sum_to_check_one_hot)
            
            # Get the labels from the one-hot encoded outputs and check the total number of correct values.
            label_pred = label_output.argmax(dim=1, keepdim=True)
            sum_pred = sum_output.argmax(dim=1, keepdim=True)
            correct += label_pred.eq(labels_to_check).sum().item()
            correct_sum += sum_pred.eq(sum_to_check).sum().item()
    
    # Get the Loss and Accuracy values.
    test_loss /= len(test_loader.dataset)
    print('\n Test Set: Loss and Accuracy of MNIST Image classification')
    print('\nTest set: Average loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
      test_loss, correct, len(test_loader.dataset),
      100. * correct / len(test_loader.dataset)))
    
    print('\n Test Set: Loss and Accuracy of MNIST Image label and Random num Sum')
    print('\nTest set: Average loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
      test_sum_loss, correct_sum, len(test_loader.dataset),
      100. * correct_sum / len(test_loader.dataset)))
```

```python
test_model()
```
